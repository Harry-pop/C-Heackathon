>***Array 1 มิติ***  ใช้ดัสนีตัวเดียวในการเข้าถึง

````python
#ตัวอย่างโปรแกรม Linear serch หาตัวเลขที่ต้องการในarray
%%writefile Array_mono.cpp
#include<iostream>

using namespace std;
int input;
int main(){
  int data[] ={1,2,43,22,322};
  int x= sizeof(data)/sizeof(data[0]) ; //หาจำนวนArrayโดยเอาขนาดarray/ขนาดข้อมูล1ตัว
  cin >> input;
  for(int i=0;i<x;) {
      if (data[i]==input){
        cout<<"Found"<<endl ;
        break;
      }else{
           cout<<"Fail"<<endl;
           i=i+1;
       }
    }

}
````

# **Record/Struct**

The differences between a class and a struct in C++ are:

โครงสร้างข้อมูล (Struct) เป็นตัวแปรที่ใช้สำหรับเก็บข้อมูลที่มีความเกี่ยวข้องกันในรูปแบบของออบเจ็ค Struct สามารถมีสมาชิกที่เป็น Primitive datatype Reference type และเมธอดได้ โครงสร้างข้อมูลนั้นคล้ายคลึงกับคลาส แต่ว่ามันจะม่สามารถสืบทอดได้ เรามักจะใช้ Struct สำหรับเก็บกลุ่มของข้อมูลที่ไม่มาก

```
Struct เป็น public ตั้งแต่แรก
Class เป็น private ตั้งแต่แรก
```
Class คือ ต้นแบบของวัตถุ (Object) สำหรับในการพัฒนาแบบ OOP (Object Oriented Programming) วัตถุแต่ละสิ่งนั้นจะมีคุณลักษณะและการทำงานที่เหมือนกัน หรือคล้ายกัน เช่น ลูกค้าต้องมีชื่อ ที่อยู่ วันเกิด และซื้อสินค้าได้ ดังนั้นเราจึงสามารถสร้างต้นแบบของลูกค้า เพื่อสะดวกในการใช้งานและแก้ไข โดยการจะนำ Class มาใช้ คือการสร้างเป็น Obect โดยใน C# จะใช้รูปแบบดังนี้




````python
%%writefile Class_OPP.cpp
#include<iostream>
using namespace std;

class med{
	public:
		int x;
		int y;
		int z;
		int sum(void){ //method
			return x+y+z;
		}
};
int main(){
	med V1;
	V1.x=8;
	V1.y=10;
	V1.z=2;
	cout<<V1.sum();
}
````

````python
! g++ Class_OPP.cpp -o Class_OPP
````

````python
! ./Class_OPP
````

การสืบทอด class คือการสร้างclassที่มีโปรแกรมเดิมมาใหม่ เป็นตัวแปรเดิมด้วย เป็นคลาสแม่สืบทอดมาเป็นลูก

````python
%%writefile Relay_Class.cpp

#include <iostream>
using namespace std;

class Parent{
	public :
	int x;
	int y;
};
class child : public Parent{
	public:
	int sum(void){
		return x+y;
	}
};
int main(){
	child exam;
	exam.x = 5;
	exam.y = 4;
	cout<<exam.sum();
}



````

````python
! g++ Relay_Class.cpp -o Relay_Class
````

````python
! ./Relay_Class
````
