````python
%%writefile Data_type.cpp

#include<iostream>
#include<typeinfo>
using namespace std;
bool x= true;
short y = 5;
int u = 50;
long j = 8;
float l = 52.2;
double g = 5.45;
const char* w ="s";
int space();
//////////////////////////////////////////////////
int main(){
	cout << "bool: " << typeid(x).name() << endl;
	cout << "short: " << typeid(y).name() << endl;
	cout << "int: " << typeid(u).name() << endl;
	cout << "long: " << typeid(j).name() << endl;
	cout << "float: " << typeid(l).name() << endl;
	cout << "double: " << typeid(g).name() << endl;
  cout << "char: " << typeid(w).name() << endl;
  	space();

}
int space(){
	cout << "space bool " << sizeof(x) << "  bit"<< endl;
	cout << "space short " << sizeof(y) << "  bit"<< endl;
	cout << "space int " << sizeof(u) << "  bit"<< endl;
	cout << "space long " << sizeof(j) << "  bit"<< endl;
	cout << "space float " << sizeof(l) << "  bit"<< endl;
	cout << "space double " << sizeof(g) << "  bit"<< endl;
	cout << "space char " << sizeof(w) << "  bit"<< endl;


}
````

````python
# Linear serch array
%%writefile Array_mono.cpp
#include<iostream>

using namespace std;
int input;
int main(){
  int data[] ={1,2,43,22,322};
  int x= sizeof(data)/sizeof(data[0]) ; //Arrayarray/1
  cin >> input;
  for(int i=0;i<x;) {
      if (data[i]==input){
        cout<<"Found"<<endl ;
        break;
      }else{
           cout<<"Fail"<<endl;
           i=i+1;
       }
    }

}
````

````python
%%writefile Class_OPP.cpp
#include<iostream>
using namespace std;

class med{
	public:
		int x;
		int y;
		int z;
		int sum(void){ //method
			return x+y+z;
		}
};
int main(){
	med V1;
	V1.x=8;
	V1.y=10;
	V1.z=2;
	cout<<V1.sum();
}
````

````python
%%writefile Relay_Class.cpp

#include <iostream>
using namespace std;

class Parent{
	public :
	int x;
	int y;
};
class child : public Parent{
	public:
	int sum(void){
		return x+y;
	}
};
int main(){
	child exam;
	exam.x = 5;
	exam.y = 4;
	cout<<exam.sum();
}



````

````python
! g++ Relay_Class.cpp -o Relay_Class
````

````python
! ./Relay_Class
````
